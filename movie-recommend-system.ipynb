# Data Visualise

```python
import pandas as pd

# Load the Excel file
file_path = 'movie_recommendation_data.xlsx'

# Load each sheet into a separate DataFrame
movies_df = pd.read_excel('/kaggle/input/movie-recommendation-system/Movies.xlsx')
users_df = pd.read_excel('/kaggle/input/movie-recommendation-system/Users.xlsx')
reviews_df = pd.read_excel('/kaggle/input/movie-recommendation-system/Reviews.xlsx')

# Display the first few rows of each DataFrame to verify the data
print("Movies DataFrame:")
print(movies_df.head())

print("\nUsers DataFrame:")
print(users_df.head())

print("\nReviews DataFrame:")
print(reviews_df.head())
```

```python
user_movie_ratings = reviews_df.pivot(index='UserID', columns='MovieID', values='Rating')
```

```python
scaler = StandardScaler()
user_movie_ratings_scaled = pd.DataFrame(scaler.fit_transform(user_movie_ratings.fillna(0)),
                                         index=user_movie_ratings.index,
                                         columns=user_movie_ratings.columns)

# Compute cosine similarity matrix
similarity_matrix = cosine_similarity(user_movie_ratings_scaled.fillna(0))
similarity_df = pd.DataFrame(similarity_matrix, index=user_movie_ratings.index, columns=user_movie_ratings.index)
```

```python
def get_recommendations(user_id, top_n=5):
    if user_id not in similarity_df.index:
        return "User ID not found in the data."

    # Get similar users
    similar_users = similarity_df[user_id].sort_values(ascending=False).index[1:]  # Exclude the user themselves

    # Aggregate ratings from similar users
    similar_users_ratings = user_movie_ratings.loc[similar_users]
    weighted_ratings = similar_users_ratings.mul(similarity_df[user_id], axis=0).sum()
    similarity_sum = similarity_df[user_id].sum()
```

# Final Movie Recommendation

```python

# Example usage
user_id = 101  # Example user ID
top_n = 5      # Number of recommendations to retrieve
recommendations = get_recommendations(user_id, top_n)
print(f"\nTop {top_n} movie recommendations for user {user_id}:A masterpiece of modern cinema")
print(recommendations)
```
